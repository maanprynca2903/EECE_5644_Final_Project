```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.util import ngrams
from collections import Counter

# Assuming 'lowercase_columns_data' is a pandas DataFrame with a column named 'complaint_text' containing preprocessed complaint text

# Concatenate all complaint texts into a single string
all_complaints_text = ' '.join(lowercase_columns_data['complaint_text'].dropna())

# Tokenize the combined text
tokens = all_complaints_text.split()

words_to_remove = ["chase", "jp", "morgan", "call", "would", "tell", "receive", "even", "though", "make", "say", "pay", "get", "send", "go", "close", "open", "save", "use", "also", "one", "days", "give", "provide", "could", "since", "file", "due", "never", "state", "ask", "back", "take", "try", "need", "know", "still", "us", "want", "day", "now", "new"]
tokens = [word for word in tokens if word.lower() not in words_to_remove]

# Count the frequency of each word
word_freq = Counter(tokens)

# Select the top 40 words by frequency
top_40_words = dict(sorted(word_freq.items(), key=lambda item: item[1], reverse=True)[:40])

# Generate and display a word cloud of the top 40 words
wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(top_40_words)

plt.figure(figsize=(9, 9))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Generate unigrams, bigrams, and trigrams
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

# Count the frequency of each n-gram
unigram_freq = Counter(unigrams)
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

# Select the top unigrams, bigrams, and trigrams by frequency
top_unigrams = unigram_freq.most_common(20)
top_bigrams = bigram_freq.most_common(20)
top_trigrams = trigram_freq.most_common(20)

# Display the top unigrams, bigrams, and trigrams
print("Top 20 Unigrams:")
for unigram, frequency in top_unigrams:
    print(' '.join(unigram), "-", frequency)

# Plotting unigrams
plt.figure(figsize=(10, 6))
plt.bar([str(uni[0]) for uni in top_unigrams], [uni[1] for uni in top_unigrams], color='skyblue')
plt.xlabel('Unigrams')
plt.ylabel('Frequency')
plt.title('Top 20 Unigrams')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nTop 20 Bigrams:")
for bigram, frequency in top_bigrams:
    print(' '.join(bigram), "-", frequency)

# Plotting bigrams
plt.figure(figsize=(10, 6))
plt.bar([str(bi[0]) for bi in top_bigrams], [bi[1] for bi in top_bigrams], color='salmon')
plt.xlabel('Bigrams')
plt.ylabel('Frequency')
plt.title('Top 20 Bigrams')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nTop 20 Trigrams:")
for trigram, frequency in top_trigrams:
    print(' '.join(trigram), "-", frequency)

# Plotting trigrams
plt.figure(figsize=(10, 6))
plt.bar([str(tri[0]) for tri in top_trigrams], [tri[1] for tri in top_trigrams], color='lightgreen')
plt.xlabel('Trigrams')
plt.ylabel('Frequency')
plt.title('Top 20 Trigrams')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(min_df=2, max_df=0.95, stop_words='english')
```

```python
doc_term_matrix = tf.fit_transform(lowercase_columns_data['complaint_text'])
doc_term_matrix
```

```python
from sklearn.decomposition import NMF

# Load your nmf_model with the n_components i.e 5
num_topics = 5

# Keep the random_state = 40
nmf_model = NMF(n_components=num_topics, random_state=40)

# Use the correct variable name
W1 = nmf_model.fit_transform(doc_term_matrix)
H1 = nmf_model.components_
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

H = nmf_model.components_       # Topic-term matrix
#Print the Top15 words for each of the topics
words = np.array(tf.get_feature_names_out())
topic_words = pd.DataFrame(np.zeros((num_topics, 15)), index=[f'Topic {i + 1}' for i in range(num_topics)],
                           columns=[f'Word {i + 1}' for i in range(15)]).astype(str)
for i in range(num_topics):
    ix = H[i].argsort()[::-1][:15]
    topic_words.iloc[i] = words[ix]

topic_words
```

```python
#Create the best topic for each complaint in terms of integer value 0,1,2,3 & 4

colnames = ["Topic" + str(i) for i in range(nmf_model.n_components)]
docnames = ["Doc" + str(i) for i in range(len(lowercase_columns_data.complaint_text))]
df_doc_topic = pd.DataFrame(np.round(W1, 2), columns=colnames, index=docnames)
significant_topic = np.argmax(df_doc_topic.values, axis=1)
df_doc_topic['dominant_topic'] = significant_topic
df_doc_topic.head()
```

```python
# utility method to get the GridSearchCV object
def grid_search(model,folds,params,scoring):

    grid_search = GridSearchCV(model,
                                cv=folds,
                                param_grid=params,
                                scoring=scoring,
                                n_jobs=-1, verbose=1)
    return grid_search
# utility function to display the best score and parameters of the GridSearchCV model
def print_best_score_params(model):
    print("Best Score: ", model.best_score_)
    print("Best Hyperparameters: ", model.best_params_)
    
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score

def plot_accuracy_threshold(model, X_test, y_test):
    # Get predicted probabilities for the positive class
    y_probs = model.predict_proba(X_test)
    
    # Assuming a binary classification for simplicity; if multiclass, select one class or adjust accordingly
    y_probs = y_probs[:, 1]  # This line assumes binary classification and selects probabilities for class 1
    
    thresholds = np.linspace(0, 1, 100)
    accuracies = []
    
    # Calculate accuracy for each threshold
    for threshold in thresholds:
        y_pred = (y_probs >= threshold).astype(int)  # Convert probabilities to binary predictions
        accuracy = accuracy_score(y_test, y_pred)
        accuracies.append(accuracy)
    
    # Plotting the accuracy curve
    plt.figure(figsize=(8, 5))
    plt.plot(thresholds, accuracies, marker='o')
    plt.title('Accuracy vs. Decision Threshold')
    plt.xlabel('Decision Threshold')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.show()

```
